```python
from random import *
from tkinter import*
from tkinter import ttk

#J'ai oublié de mettre a jour sur github la semaine de dernière mais voilà un résumer de ce que j'ai fait depuis la dernière version
#ajout de l'algorithme de dégagement
#ajout des 3 mode de difficulté avec leurs différents paramètre 
#ajout de la liste des niveau et du bouton qui sert a lancer la partie (fonctionnel)



def afficher(M):
    "Affiche une matrice en respectant les alignements par colonnes"
    w=[max([len(str(M[i][j])) for i in range(len(M))]) for j in range(len(M[0]))]
    for i in range(len(M)):
        for j in range(len(M[0])):
            print("%*s" %(w[j],str(M[i][j])), end= ' ')
        print()    

def voisins(n, i, j):
    """Dans une grille n x n, indices (ligne, colonne) des voisins 
    dans la grille d'un sommet (i,j) de la grille"""
    return [(a,b) for (a, b) in [(i,j+1),(i,j-1),(i+1,j),(i-1,j),(i+1,j+1),(i+1,j-1),(i-1,j+1),(i-1,j-1)] if a in range(n) and b in range(n)]

def bomb(L,n,l,c):
    vbomb=voisins(n,l,c)#liste des voisions d'une bombe
    vbomb_len=len(vbomb)
    for i in range(len(L)):
        for j in range(len(L[0])):
            for (a, b) in vbomb:
                if (a,b)==(i,j): #parcours les coordonées de L et les compare avec celle dans vbomb et si elles sont egale
                    x=L[i][j]    #x =x+1
                    if x!=9:
                        x=x+1
                    L[i][j]=x



def jeu(event):   # quand on cliqe sur une case ça renvoit les coordonées correspondant à la liste L
    y=event.x//unit
    x=event.y//unit
    if L[x][y]!=9 and L[x][y]!= 0:
        x=L[x][y]
        a=(event.x//unit)*unit
        b=(event.y//unit)*unit
        cnv.create_text(a+unit//2, b+unit//2, text=x, font=("Arial 50 bold",10))
    elif L[x][y]==0: # donc qu'il n'y a pas de bombe on peut lancer le dégagement
        cnv.create_text(((y*unit)+unit//2), ((x*unit)+unit//2), text=L[x][y], font=("Arial 50 bold",10))
        dégagement(x,y)
    else:
        print("bombe !")
        for i in range(len(L)):
            for j in range(len(L[0])):
                if L[i][j]==9:
                    a=j*50
                    b=i*50
                    cnv.create_oval((10+a,10+b,a+40,b+40),fill='black',width=1,outline="black") # montre toute les bombe quand on clic sur UNE bombe
                

                
nbrflag=0
F=[]
O=[]
def flag (event):
    global nbrflag
    global b
    y=event.x//unit
    x=event.y//unit
    a,b=y*unit,x*unit
    if nbrflag<b and (a,b) not in F: # le "nbrflag" pour limiter le nombre de drapeau que l'on peux poser au nombre de bombe .
        # et la liste F permet de noté si un flag a déja etait mis à un cet endroit 
        O.append(cnv.create_oval((10+a,10+b,a+40,b+40),fill='red',width=1,outline="red"))#crée l'oval et met son id dans la liste O
        nbrflag=nbrflag+1
        F.append((a,b))
    elif (a,b) in F:
        q=0
        p=True
        while p :
            if (a,b) == F[q]:
                p=False
            else:
                q=q+1
        #q Est l'emplacement dans la liste de l'id de l'oval à supprimer 
        cnv.delete(O[q])
        del O[q]
        del F[q]
        nbrflag=nbrflag-1

            
I=[] #emplacement des ?
idI=[] #id des ?
def interro (event):#double clic droit pour mettre un point d'interrogation
    global n
    y=event.x//unit
    x=event.y//unit
    a,b=y*50,x*unit
    if (a,b) not in I: #On fait pareil qu'avec les ! sauf qu'on s'en fiche du nombre (car ilimité)
        idI.append(cnv.create_oval((10+a,10+b,a+40,b+40),fill='green',width=1,outline="green"))
        I.append((a,b))
    elif (a,b) in I:
        q=0
        p=True
        while p :
            if (a,b) == I[q]:
                p=False
            else:
                q=q+1
        cnv.delete(idI[q])
        del idI[q]
        del I[q]





            
CASEVIDE=[]      
def dégagement(x,y): #cette fonction est utilisée si sur la case cliquée il n'y a pas de voisins bombe et va propager ce clic dans tous les sens  jusqu'a qu'il arrive sur un case qui à au moins une bombe voisines 
    Q=voisins(n,x,y)
    for i in Q:
        (x,y)=i
        if (x,y) not in CASEVIDE: #on verifie que un case n'est pas déja était afficher 
            cnv.create_text(((y*unit)+unit//2), ((x*unit)+unit//2), text=L[x][y], font=("Arial 50 bold",10)) 
            CASEVIDE.append((x,y))
            if L[x][y]==0:
                dégagement(x,y)

        
             

def Menu(event):
    global n
    global n2
    global L
    global b
    global unit
    select = listeCombo.get()
    if select=="Facile":
        print("Facile")
        unit=50
        n=8
        n2=8
        L=[[],[],[],[],[],[],[],[]]
        b=10
        for i in range (len(L)):
            for j in range (n):
                 L[i].append(0)
        
        for p in range (b):
            l=randrange(n)
            c=randrange(n)
            L[l][c]=9
            bomb(L,n,l,c)
        afficher(L)
        


        return n,n2,b,L,unit
    elif select=="Normal":
        print("Normal")
        unit=50
        n=16
        n2=16
        b=40
        L=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
        for i in range (len(L)):
            for j in range (n):
                 L[i].append(0)
        
        for p in range (b):
            l=randrange(n)
            c=randrange(n)
            L[l][c]=9
            bomb(L,n,l,c)
        afficher(L)
        

        return n,n2,b,L,unit
    elif select=="Difficile":
        print("Difficile")
        unit=25
        n=32
        n2=16#ligne
        b=99
        L=[[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
        for i in range (len(L)):
            for j in range (n):
                 L[i].append(0)
        
        for p in range (b):
            l=randrange(n2)
            c=randrange(n)
            L[l][c]=9
            bomb(L,n,l,c)
        afficher(L)

        return n,n2,b,L,unit




root=Tk()


cnv = Canvas(root, width=900, height=700, background="ivory") 
cnv.pack(padx=10, pady=10, side=LEFT)

def jouer():
    cnv.delete("all")
    init()
    
def init(): # initialise les cases du jeu.
    for line in range(n2):
        for col in range(n):
            A=(unit*col, unit*line)
            B=(unit*(col+1), unit*(line+1))
               
            cnv.create_rectangle(A, B, fill="grey",)
    



l1=Label(root,text="Choix de la difficulté")
l1.pack()
listeProduits=["Facile", "Normal","Difficile"]
listeCombo = ttk.Combobox(root,text= "Choix de la difficulté", values=listeProduits)
listeCombo.bind("<<ComboboxSelected>>", Menu)
listeCombo.pack()               
                            

cnv.bind("<Button-1>",jeu)
cnv.bind("<Button-3>",flag)
cnv.bind("<Double-Button-3>",interro)
btn = Button(root, text="Lancer la partie ", command=jouer)
btn.pack()


    
   
                

root.mainloop()
```
